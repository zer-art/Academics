# MONTE CARLO SIMULATION #



**importing modules**

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import random
```

**creating axis and variables**


```python
x_axis = []
y_axis = []
inside = 0
```

**taking user input**

```python
print(" 1 = 100 \n 2 = 1000 \n 3 = 10000 \4 = custom ")
while True:
    user_input = int(input("enter number of terms "))
    if user_input == 1 : 
       terms = 100
       break
    elif user_input ==2 : 
        terms =1000
        break
    elif user_input == 3 : 
        terms = 10000
        break
    elif user_input == 4 :
        terms = input("enter num : ")
        try : 
            terms = int(terms)
            break
        except ValueError : 
            print(f"{terms} is not a calid number .") 
    else:
        print("select a valid option : ")   
```

**coding logic**


```python
for i in range (terms) :
    x = random.uniform(0,1)
    y = random.uniform(0,1)
    if x**2 + y**2 <= 2**1/2 :
        x_axis.append(x)
        y_axis.append(y)  
        inside +=1 
    
pi = (inside/terms) * 4
print(f"estimated value of pi is : {pi}")          
```

**representation**


```python
in_area = sns.scatterplot(x=x_axis,y=y_axis)
```
