**Importing liberaries**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

**Ploting and generatig normalised data having mean = 5 and std = 10**

```python
data = np.random.normal(5, 10, 1000)  

print(f"mean of data {data.mean()}")
print(f"mean of data {data.std()}")

plt.hist(data, bins=100 , edgecolor='k') 
plt.title('Histogram of Normalized Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Ploting price over time data of TCS**

```python
df = pd.read_csv('Quote-Equity-TCS-EQ-28-06-2023-to-28-06-2024.csv')
df.head()
```

```python
plt.plot(df["Date "], df['OPEN '], linestyle='-',label = "VALUE ")
plt.title('Price Over Time In TCS')
plt.xlabel('Date')
plt.xticks([])
plt.yticks([])
plt.ylabel('Price')


plt.legend()
plt.show()
```

```python
plt.bar(df["Date "], df['VALUE '], label='Price')
plt.title('Price Over Time In TCS')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks([])
plt.yticks([])
plt.legend()
plt.show()
```
