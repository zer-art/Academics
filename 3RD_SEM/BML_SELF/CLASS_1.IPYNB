**Machine learning Supervised Regression**

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets , linear_model 
from sklearn.metrics import mean_absolute_percentage_error
```

```python
diabetes = datasets.load_diabetes()

# all attributes present in the datasets 
print(diabetes.keys()) 

#dataset description 
diabetes.DESCR
```

```python
diabetes_x = diabetes.data[:, np.newaxis , 2 ] # consists independent variables 
diabetes_y = diabetes.target[:] # consists dependent variable 
diabetes_x ,diabetes_y
```

```python
#using data of last 30 indices for training the model 
diabetes_x_train = diabetes_x[:-30]
diabetes_y_train = diabetes_y[:-30]
```

```python
#using data of first 20  indices for testing the model 
diabetes_x_test = diabetes_x[:20]
```

*Generating model*

```python
model = linear_model.LinearRegression()
model.fit(diabetes_x_train,diabetes_y_train)
```

*Getting model output*

```python
diabetes_y_predicted = model.predict(diabetes_x_test)
```

```python
diabetes_y_actual = diabetes_y[:20] # actual value of dependent variable 

#creating a function to calculate mean square error 
error_array = diabetes_y_actual**2 - diabetes_y_predicted**2
mean_squared_error = np.mean(error_array)
mean_squared_error
```
