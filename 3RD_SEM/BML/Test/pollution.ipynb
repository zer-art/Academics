```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# i am using this sample kaggle dataset just replace it with your original file 
df = pd.read_csv('air_pollution_data.csv')
df.head(10)
```

```python
overview = df.describe() 
overview
```

```python
# remove this if yoou dont need 
# i am removing unnecessary rows from the data 
df_requied=df.loc[df['city'].str.lower() == 'ahmedabad', [ 'date', 'pm10', 'pm2_5']]
df_requied.head()
```

```python
overview_df_requied = df_requied.describe() 
overview_df_requied
```
