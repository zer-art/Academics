# SEABORN LIBRARY

```python
import seaborn as sns
from matplotlib import pyplot as plt 
```

***LOAD DATA***

```python
fmri = sns.load_dataset("fmri")
penguins = sns.load_dataset("penguins")
```

***LINEPLOT***

```python
plot_1 = sns.lineplot(x="timepoint",y = "signal",hue="region" ,style= "event" , data=fmri ,legend = None)
```

***HISTPLOT***

```python
plot_2 = sns.histplot(data=penguins , x="body_mass_g")
```
