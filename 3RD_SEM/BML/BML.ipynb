# LEARNING PYPLOT

```python
xpoint  =np.array([0,6,4])
ypoint = np.array([0,250,0])

plt.plot(xpoint,ypoint , marker = "o" , linestyle = "--" , ms = 10 , mec = "b")
plt.xlabel("datpoints")
plt.ylabel("attributes")
plt.title("graph")
plt.show()
```

# LEARNING BAR PLOTTING

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([31, 18, 11, 20])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([31, 18, 11, 20])

plt.barh(x,y)
plt.show()
```
